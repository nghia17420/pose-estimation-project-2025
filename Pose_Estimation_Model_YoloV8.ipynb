#1/ TEST ON COCO DATASET


```python
#COCO pretrained pose model (small size)
model = YOLO('yolov8s-pose.pt')

#upload video to test model
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
cap = cv2.VideoCapture(video_path)

#get video properties for output video
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#fourcc = four characters code specify the video codec
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('cocoYoloV8sTest.mp4', fourcc, fps, (width, height))

#chop the video into frames to run pose estimation
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read() #read next video frame. ret is True if the frame was read successfully
    if not ret:
        break
    #pose inference on current frame
    results = model(frame)
    #draw predicted keypoints and skeleton on frame
    annotated_frame = results[0].plot()
    # Write frame to output video
    out.write(annotated_frame)
    frame_count += 1
    if frame_count % 30 == 0:
        print(f'Processed {frame_count} frames...')

cap.release()
out.release()

#download test output
files.download('cocoYoloV8sTest.mp4')
```
