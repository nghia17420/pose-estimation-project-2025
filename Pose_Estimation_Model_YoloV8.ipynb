```python
from ultralytics import YOLO

#load pretrained model on COCO dataset
model = YOLO('yolov8n-pose.pt')

# Run inference on a deadlift image/video
results = model('path/to/deadlift_image.jpg')

# Visualize
for r in results:
    im_array = r.plot()
    Image.fromarray(im_array[..., ::-1]).show()
```
