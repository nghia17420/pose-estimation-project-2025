```python
#install required libraries
!pip install ultralytics opencv-python

from ultralytics import YOLO
import cv2
```

```python
# Load pretrained model on COCO dataset
model = YOLO('yolov8n-pose.pt')

# Uncomment to train on custom data
# model.train(data='deadlift_pose.yaml', epochs=100, imgsz=640, batch=16)

# Validate performance
metrics = model.val()
print(f"mAP50-95: {metrics.pose.map}")
```

```python
# Export to ONNX with Hailo-compatible settings
model.export(format='onnx', opset=11, imgsz=640, simplify=True)

# Download from Colab
from google.colab import files
files.download('yolov8n-pose.onnx')
```
